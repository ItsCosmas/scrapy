```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
i = 1
dict_list = []
while True:
    base_url = "https://retro.ihub.co.ke/jobs/"+str(i)
    res = requests.get(base_url)
    c = res.content
    soup = BeautifulSoup(c, "html.parser")
    
    all_jobs_rows = soup.find_all("div", {"class": "jobsboard-row"})
        
    for item in all_jobs_rows:
        dict = {}
        dict["category"]=item.find("div",{"class":"job-cat"}).text
        dict["company"]=item.find("a",{"class":"post-company"}).text
        dict["date_posted"]=item.find("div",{"class":"job-time"}).text
        dict["desc"]=item.find("div",{"class":"post-description"}).text.replace("\n","").replace(u'\xa0', u'') 
        dict["link"]='https://ihub.co.ke' + item.find("a",{"class":"job-more"}, href=True)['href']
        dict["location"]=item.find("div",{"class":"job-location"}).text
        dict["title"]=item.find("h3").text.replace("\n","")


        dict_list.append(dict)
    
    if soup.find("li", {"class": "next"}) != None:
        i = i + 1
    else:
        break
    
#     delay for 2 seconds to crawl responsibly
    time.sleep(2)
```

```python
len(dict_list)
```

```python
df = pd.DataFrame(dict_list)
```
